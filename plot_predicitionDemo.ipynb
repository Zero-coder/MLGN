```python
import torch.nn.functional as F
import torch
import matplotlib.pyplot as plt 
import numpy as np
import matplotlib
import random
```

```python
true = np.load('true.npy')
x = np.load('x.npy')

print(true.shape)
print(x.shape)

true = torch.from_numpy(true).permute(0,2,1)
x = torch.from_numpy(x).permute(0,2,1)

print(true.shape)
print(x.shape)
#取 第i个batch 的最后OT列
i=0

true_sample = true[i][-1]
x_sample = x[i][-1]

print(true_sample.shape)
print(x_sample.shape)
```

```python
y = np.array(true_sample)
y[:300] = true_sample[36:336]
for i in range(36):
    y[300+i] = y[300+i-1] - random.uniform(0,0.05)
```
